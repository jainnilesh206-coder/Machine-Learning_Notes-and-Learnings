```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv')
```

```python
data.info()
```

```python
features = data.iloc[:,[0,1,2,3]].values
label = data.iloc[:,[4]].values
```

```python
#Perform Feature Reduction using PCA
# 1. Apply Standardization on feature columns ---- (StandardScaler)
# 2. Apply PCA on the Feature Column
```

```python
# 1, Apply StandardScaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaledFeatures = sc.fit_transform(features)
```

```python
# 2. Train PCA object

from sklearn.decomposition import PCA
pc = PCA(n_components=2) #n_components must not be greater than no of feature columns | n >= 1 & n <= noOfFeatures
pc.fit(scaledFeatures)
```

```python
# 3. Extract Principal Components

finalFeatureSet = pc.transform(scaledFeatures)
```

```python
finalFeatureSet
```

```python
#Modelling Start

# Create Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(finalFeatureSet,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#Deployment

sepalLength = float(input("Enter sepal Length: "))
sepalWidth = float(input("Enter sepal width: "))
petalLength = float(input("Enter petal length: "))
petalWidth = float(input("Enter petal width: "))

#feature set

featureSet = np.array([[sepalLength,sepalWidth, petalLength, petalWidth]])

#Apply SC object

stdFeatures = sc.transform(featureSet)

#Apply PCA object

transformedFeatures = pc.transform(stdFeatures)

#Pass transforedFeatures to Model

model.predict(transformedFeatures)
```
