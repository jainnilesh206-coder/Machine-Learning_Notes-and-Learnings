```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
```

```python
data.info()
```

```python
features = data.iloc[:,[0,1]].values
label = data.iloc[:,[2]].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=24)
```

```python

from sklearn.ensemble import RandomForestClassifier #For Classification
#from sklearn.ensemble import RandomForestRegressor #for Regression

ensembleModel = RandomForestClassifier(
    n_estimators=5,
    random_state=2,
    max_depth=5

)
```

```python
ensembleModel.fit(X_train,y_train)
```

```python
ensembleModel.score(X_test,y_test)

#In sklearn all classification algo--- default metric for score() is accuracy
# all regression algo --- default metric for score() is R2score
```

```python
ensembleModel.score(X_train,y_train)
```

```python
# Sampling --- full datasample
import warnings
warnings.filterwarnings('ignore')
CL = 0.95


for rs in range(101):
  for depth in range(3,11):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=rs)
    ensembleModel = RandomForestClassifier(
    n_estimators=5,
    random_state=2,
    max_depth=depth)

    ensembleModel.fit(X_train,y_train)
    testScore = ensembleModel.score(X_test,y_test)
    trainScore = ensembleModel.score(X_train,y_train)

    if testScore > trainScore and testScore > CL:
      print(f"Test Score {testScore} and TrainScore {trainScore} for depth {depth} and rs {rs}")
```

```python

import warnings
warnings.filterwarnings('ignore')
CL = 0.95

sampleSize = int(round(np.sqrt(len(X_train))))


for rs in range(101):
  for depth in range(3,11):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=rs)
    ensembleModel = RandomForestClassifier(
    n_estimators=5,
    random_state=2,
    max_depth=depth,
    max_samples=sampleSize)

    ensembleModel.fit(X_train,y_train)
    testScore = ensembleModel.score(X_test,y_test)
    trainScore = ensembleModel.score(X_train,y_train)

    if testScore > trainScore and testScore > CL:
      print(f"Test Score {testScore} and TrainScore {trainScore} for depth {depth} and rs {rs}")
```

```python
#Shuffling
import warnings
warnings.filterwarnings('ignore')
CL = 0.95



for rs in range(101):
  for depth in range(3,11):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=rs)
    ensembleModel = RandomForestClassifier(
    n_estimators=5,
    random_state=2,
    max_depth=depth,
    bootstrap=False)

    ensembleModel.fit(X_train,y_train)
    testScore = ensembleModel.score(X_test,y_test)
    trainScore = ensembleModel.score(X_train,y_train)

    if testScore > trainScore and testScore > CL:
      print(f"Test Score {testScore} and TrainScore {trainScore} for depth {depth} and rs {rs}")
```
