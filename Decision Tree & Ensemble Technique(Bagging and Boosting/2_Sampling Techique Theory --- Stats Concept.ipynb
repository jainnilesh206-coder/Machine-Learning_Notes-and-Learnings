```python
import numpy as np
```

```python
data = np.array([1,2,3,4,5,6])
```

# Sampling with Replacement

```python
sample = np.random.choice(data, size=39, replace=True)
sample
```

# Sampling without Replacement

```python
sample = np.random.choice(data, size=3, replace=False)
sample
```

```python
sample = np.random.choice(data, size=30, replace=False)
sample
```
