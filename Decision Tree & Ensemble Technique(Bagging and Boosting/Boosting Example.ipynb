```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cancer.csv")
```

```python
features = data.iloc[:,2:32].values
label = data.iloc[:,[1]].values
```

```python
# pip install xgboost
```

```python
# Expectation/Rule:
# 1. XGBoost expects label column to be represented in Numerical way
#        Classification --- Discrete Numerical
#        Regression ------- You are good (Continuous ND)
```

```python
label
```

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(label)
```

```python
label
```

```python
le.transform(np.array([['B']]))
```

```python
le.inverse_transform(np.array([1]))
```

```python
# XGBClassifier -- DTalgo as a estimator

CL=0.8

from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

for i in range(1,301):
  X_train,X_test,y_train,y_test = train_test_split(features,
                                                   label,
                                                   test_size=0.2,
                                                   random_state=i)

  model = XGBClassifier()
  model.fit(X_train,y_train)
  testScore = model.score(X_test,y_test)
  trainScore = model.score(X_train,y_train)

  if testScore>trainScore and testScore > CL or testScore == trainScore:
    print(f"Test Score : {testScore} | trainScore: {trainScore} | rs: {i}")
```

```python
# XGBClassifier -- DTalgo as a estimator

CL=0.8

from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

for i in range(1,301):
  for md in range(3,11):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                   label,
                                                   test_size=0.2,
                                                   random_state=i)

    model = XGBClassifier(max_depth=md)
    model.fit(X_train,y_train)
    testScore = model.score(X_test,y_test)
    trainScore = model.score(X_train,y_train)

    if testScore>trainScore and testScore > CL:
      print(f"Test Score : {testScore} | trainScore: {trainScore} | rs: {i} | md {md}")
```

```python
# XGBClassifier -- DTalgo as a estimator

CL=0.8

from sklearn.model_selection import train_test_split
from xgboost import XGBRFClassifier

for i in range(1,51):
  X_train,X_test,y_train,y_test = train_test_split(features,
                                                   label,
                                                   test_size=0.2,
                                                   random_state=i)

  model = XGBRFClassifier()
  model.fit(X_train,y_train)
  testScore = model.score(X_test,y_test)
  trainScore = model.score(X_train,y_train)

  if testScore>trainScore and testScore > CL or testScore == trainScore:
    print(f"Test Score : {testScore} | trainScore: {trainScore} | rs: {i}")
```
