```python
# KNN ---
# - Supervised learning Algorithm (applicable for both Regression and Classification)
# - sklearn implementation
#        - Regression: KNeighborsRegressor
#        - Classification: KNeighborsClassifier
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data.info()
```

```python
data.Purchased.unique()
```

```python
data.Purchased.value_counts()
```

```python
# assume
# 0 -------- Bad Customer (Customer will not make a purchase)
# 1 -------- Good Customer (Customer will make a purchase)
```

```python
# Seeprate data as features and label
features = data.iloc[:,[0,1]].values
label = data.iloc[:,[2]].values
```

```python
CL= 0.8

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

for rs in range(1,301):
  X_train,X_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state=rs)

  for k in range(3,11):
    model = KNeighborsClassifier(n_neighbors=k) # n_neighbors ----> k
    model.fit(X_train,y_train)

    trainScore = model.score(X_train,y_train)
    testScore = model.score(X_test,y_test)

    if testScore > trainScore and testScore >= CL:
      print(f"Test Score : {testScore} | Train Score : {trainScore} | rs: {rs} | k: {k}")
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,label,test_size=0.2,random_state=116)


model = KNeighborsClassifier(n_neighbors=7) # n_neighbors ----> k
model.fit(X_train,y_train)

trainScore = model.score(X_train,y_train)
testScore = model.score(X_test,y_test)

print(f"Test Score : {testScore} | TrainScore: {trainScore}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(label, model.predict(features)))
```
