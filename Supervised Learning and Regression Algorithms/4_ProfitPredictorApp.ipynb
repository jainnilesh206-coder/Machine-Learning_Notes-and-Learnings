```python
import pickle
model = pickle.load(open("ProfitPredictor.pkl",'rb'))
oheState = pickle.load(open("StateEncoder.pkl",'rb'))
```

```python
#Deploy app
import numpy as np

rdSpend = float(input("Enter RDSpend: "))
adminSpend = float(input("Enter AdminSpend: "))
marketSpend = float(input("Enter MarketSpend: "))
state = input("Enter State: ")


#Check if the state is valid state for the ml model

if state in oheState.categories_[0]:

  dummyState = oheState.transform(np.array([[state]]))

  finalFeatureInput = np.concatenate( (dummyState,np.array([[rdSpend,adminSpend,marketSpend]])) , axis = 1)

  profit = model.predict(finalFeatureInput)

  print(f"Profit is {profit}")

else:
  print(f"Given State {state} is not recognized by AI model")
```
