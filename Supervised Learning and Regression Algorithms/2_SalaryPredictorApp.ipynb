```python
import pickle
import numpy as np
```

```python
model = pickle.load(open('SalaryPredictor-2.pkl', 'rb'))
```

```python
# Deploy Model ----- Help the Dev Team to understand how to use the trained model
#

yExperience = float(input("Enter years of experience: "))

predictedSalary = model.predict(np.array([[yExperience]]))

print(f"Salary for {yExperience} years of experience is ${predictedSalary}")
```
