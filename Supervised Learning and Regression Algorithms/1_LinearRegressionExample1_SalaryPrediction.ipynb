```python
# Create a model for an HR company that can predict the salary of the employee based
# on his/her years of experience
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Salary_Data.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
#Deleting a record based on index (DYnamic)
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# Rules for Regression
# 1. Data must be COMPLETE
# 2. Data must be STRICTLY NUMERIC
# 3. Feature and Label must be represented as 2d np array
```

```python
#Step1: Seperate data as features and label

features = data.iloc[:,[0]].values
label = data.iloc[:,[1]].values
```

```python
# Step2: Create Train Test Split

#In sklearn, we have a function named train_test_split()
#
# train_test_split return 4 variables
# X_train ------ training feature set
# X_test ------- testing feature set
# y_train ------ training label set
# y_test ------- testing label set

# training data ---> (X_train, y_train)
# testing data ----> (X_test, y_test)

# Usually train test split ratio --- depends on data size
# 90:10 split --- -small data
# 80:20 split --- standard size
# 65:35 split --- large data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size=0.2,
                                                    random_state= 10) #Shuffles the data ---- shuffle seed
```

```python
X_train
```

```python
# Step3: Select the ML algo ------ LinearRegression

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train) #Fit() is responsible to find the coeff and intecept


# salary = b0 + b1(yExp)
```

```python
#b0
model.intercept_
```

```python
#b1
model.coef_
```

```python
# salary = 26089 + 9356 (yExp)
```

```python
model.predict(np.array([[0]]))
```

```python
#
#
# Check the quality of the model
#
# SL (Significance Level) ---------------------- 0.05
#
# CL = 0.95
#
# Model must achieve atleast 95% score
```

```python
#
# When it comes to approving a model, always go for generalized model
#
# Generalized Model(Rancho) means the trained model understands the POPULATION'S PATTERN.
# This model not only performs BEST in known data, but also works BEST with unknown data.
#
#
# =========================================================================================
# Guideline by Prashant Nair to identify the generalized model
# =========================================================================================
#
# a. Calc the evaluation score of the model w.r.t training data nad testing data
#
#     trainScore ------- Score generated by Training data
#     testScore -------- Score generated by Testing data
#
# b. Check the below criteria
#
#     if testScore > trainScore and testScore >= CL:
#                   Approve model (Generalized model)
#     else:
#                   Improve the model | Start Optimization Process
```

```python
trainScore = model.score(X_train,y_train)
testScore = model.score(X_test,y_test)

print(f"Test Score is {testScore} and train score is {trainScore}")
```

```python
# Statistics GOLDEN RULE ---- " The way how you present the data, INFLUENCE the learning power of AI algorithm "

# My own Dr. AI Strange

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

CL = 0.95
win =0

for rs in range(1,101):
  X_train,X_test,y_train,y_test = train_test_split(features,
                                                   label,
                                                   test_size=0.2,
                                                   random_state=rs)

  model = LinearRegression()
  model.fit(X_train,y_train)

  trainScore = model.score(X_train,y_train)
  testScore = model.score(X_test,y_test)

  if testScore > trainScore and testScore >=CL:
    win+=1
    print(f"TestScore: {testScore} | trainScore: {trainScore} | rs: {rs}")

print(f"\nTotal Wins : {win} ")
```

```python
# Deploy Model ----- Help the Dev Team to understand how to use the trained model
#

yExperience = float(input("Enter years of experience: "))

predictedSalary = model.predict(np.array([[yExperience]]))

print(f"Salary for {yExperience} years of experience is ${predictedSalary}")
```
